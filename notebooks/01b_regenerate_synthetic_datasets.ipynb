````python
def validate_and_integrate(plain_llm_json_str, hybrid_vtriad_json_str):
    """
    Validate JSON format and integrate with existing datasets
    """
    try:
        plain_data = json.loads(plain_llm_json_str)
        hybrid_data = json.loads(hybrid_vtriad_json_str)
    except json.JSONDecodeError as e:
        print(f"❌ JSON Parse Error: {e}")
        return None
    
    print(f"✓ Plain LLM: {len(plain_data)} emails")
    print(f"✓ Hybrid V-Triad: {len(hybrid_data)} emails")
    
    # Validate structure
    required_fields = ['subject', 'sender', 'body']
    
    for dataset_name, dataset in [('Plain LLM', plain_data), ('Hybrid', hybrid_data)]:
        for i, email in enumerate(dataset):
            for field in required_fields:
                if field not in email:
                    print(f"❌ {dataset_name}[{i}] missing field: {field}")
                    return None
            if not isinstance(email.get('urls', []), list):
                print(f"❌ {dataset_name}[{i}] urls must be a list")
                return None
    
    print("\n✓ All validation checks passed!")
    return {'plain_llm': plain_data, 'hybrid_vtriad': hybrid_data}

# Call validation when ready (after pasting JSON)
# results = validate_and_integrate(plain_llm_json, hybrid_vtriad_json)
````

## 4. Automatic Generation (Programmatic)

This section generates sophisticated emails programmatically using audit insights:
- Institutional contexts (universities, companies)
- Subtle urgency (not obvious threats)
- Professional tone (perfect grammar)
- Realistic URLs (legitimate-looking domains)
- No generic patterns that trigger simple detection

````python
# Generation templates based on audit findings
import random
random.seed(42)

# Real institutional contexts to use
institutions = [
    {'name': 'Cornell University', 'domain': 'cornell.edu', 'system': 'NetID', 'dept': 'IT Services'},
    {'name': 'Stanford University', 'domain': 'stanford.edu', 'system': 'SUNet ID', 'dept': 'Information Technology'},
    {'name': 'MIT', 'domain': 'mit.edu', 'system': 'Kerberos', 'dept': 'IS&T'},
    {'name': 'UC Berkeley', 'domain': 'berkeley.edu', 'system': 'CalNet ID', 'dept': 'IT Services'},
    {'name': 'Harvard University', 'domain': 'harvard.edu', 'system': 'Harvard Key', 'dept': 'HUIT'},
]

companies = [
    {'name': 'TechCorp', 'domain': 'techcorp.com', 'dept': 'IT Security', 'system': 'SSO'},
    {'name': 'DataFlow Inc', 'domain': 'dataflow.com', 'dept': 'Security Operations', 'system': 'Identity'},
    {'name': 'CloudVenture', 'domain': 'cloudventure.com', 'dept': 'IT Operations', 'system': 'Access Management'},
    {'name': 'FinServe', 'domain': 'finserve.com', 'dept': 'Compliance & Security', 'system': 'Authentication'},
]

# Realistic HR/IT processes (not obvious threats)
# Avoid account/profile/identity/credentials language
processes = [
    {'type': 'benefits', 'name': 'Open Enrollment', 'urgency': 'Please complete by Friday', 'action': 'review the benefits summary'},
    {'type': '2fa', 'name': '2-Factor Authentication', 'urgency': 'Please review this week', 'action': 'review the guidance document'},
    {'type': 'policy', 'name': 'Security Policy Update', 'urgency': 'Please acknowledge', 'action': 'review the updated policy notice'},
    {'type': 'training', 'name': 'Compliance Training', 'urgency': 'Due by end of month', 'action': 'review the training outline'},
    {'type': 'audit', 'name': 'Annual Compliance Review', 'urgency': 'Routine review period', 'action': 'review the compliance memo'},
]

print("✓ Generation templates loaded")
print(f"✓ Institutional contexts: {len(institutions)} universities + {len(companies)} companies")
print(f"✓ Processes: {len(processes)} types")
````

````python
def generate_plain_llm_emails(count=50):
    """
    Generate Plain LLM phishing emails.
    Characteristics (from audit):
    - Subtle, context-appropriate urgency
    - Professional tone, no obvious patterns
    - Institutional processes (not generic)
    - No URLs (QR/attachment-only references)
    - Avoid account/profile/identity/credentials language
    """
    emails = []
    
    for i in range(count):
        # Mix of institutions and companies
        if i % 2 == 0:
            org = random.choice(institutions)
            sender = f"{random.choice(['info', 'noreply', 'support'])}@{org['domain']}"
        else:
            org = random.choice(companies)
            sender = f"{random.choice(['admin', 'noreply', 'hr'])}@{org['domain']}"
        
        process = random.choice(processes)
        
        # Subject: Professional, no obvious red flags
        subjects = [
            f"{process['name']} - {process['urgency']}",
            f"Update: {process['name']}",
            f"Please Review: {process['name']}",
            f"{process['name']} - Important Notice",
        ]
        subject = random.choice(subjects)
        
        # Body: QR/attachment-only references, no access language
        body = f"""Hello,

As part of our standard operations, we are conducting a {process['name'].lower()} for {org['name']} personnel.

{process['urgency']}. The reference materials are provided as an attached PDF and a QR code in the memo.

Please review the memo at your convenience. If you believe you received this message in error, contact {org['dept']} at support@{org['domain']}.

Best regards,
{org['dept']} Team
{org['name']}"""
        
        emails.append({
            'subject': subject,
            'sender': sender,
            'body': body,
            'urls': []
        })
    
    return emails

# Generate Plain LLM
plain_llm_emails = generate_plain_llm_emails(50)
print(f"\n✓ Generated {len(plain_llm_emails)} Plain LLM emails")
print(f"  Sample sender: {plain_llm_emails[0]['sender']}")
print(f"  Sample subject: {plain_llm_emails[0]['subject']}")
````

````python
def generate_hybrid_vtriad_emails(count=50):
    """
    Generate Hybrid V-Triad phishing emails (most sophisticated).
    Characteristics (from audit):
    - Multi-factor social engineering (3+ cues per email)
    - Authority impersonation (fake internal titles)
    - Real processes + realistic urgency
    - Longer, more detailed (400+ words)
    - No obvious red flags
    - No URLs (QR/attachment-only references)
    - Avoid account/profile/identity/credentials language
    """
    emails = []
    
    titles = [
        'IT Security Team', 'Compliance Officer', 'System Administrator',
        'Security Operations Center', 'Identity & Access Management',
        'Information Security Office', 'Network Security Division'
    ]
    
    for i in range(count):
        org = random.choice(institutions + companies)
        sender_name = random.choice(['John', 'Sarah', 'Michael', 'Lisa', 'Robert'])
        sender_title = random.choice(titles)
        sender = f"{sender_name.lower()}.security@{org['domain']}"
        
        process = random.choice(processes)
        
        # Subject: Professional but subtle urgency (avoid URGENT/ALERT keywords)
        subjects = [
            f"{process['name']} - {org['name']} Update",
            f"Required Review: {process['name']}",
            f"{org['system']} Maintenance Notice",
            f"{process['name']} Review - {org['dept']}",
        ]
        subject = random.choice(subjects)
        
        # Body: Long, detailed, multi-factor social engineering
        body = f"""Hello {org['name']} Colleague,

This is a formal notice from the {sender_title} regarding a scheduled {process['name'].lower()} across the organization.

CONTEXT:
As part of our ongoing operational review, all teams are asked to complete a brief review of current records and policy updates related to {org['system']} services. This supports the annual compliance cycle and aligns with internal audit timelines.

WHAT YOU NEED TO DO:
1. Review the materials provided in the attached memo
2. Confirm that your team has acknowledged the notice internally
3. Ensure departmental records are current for this review cycle

REFERENCE MATERIAL:
- Attached PDF memo
- QR code included in the memo for reference

TIMELINE:
{process['urgency']}. If your team has already completed this review, you can disregard this notice.

ADDITIONAL NOTES:
- This review is aligned with standard organizational procedures
- No sensitive information should be sent via email
- The memo provides all necessary documents and guidance

If you have questions, contact {org['dept']} Help Desk at support@{org['domain']}.

Regards,

{sender_name} {sender_title}
{org['name']} - {org['dept']}
"""
        
        emails.append({
            'subject': subject,
            'sender': sender,
            'body': body,
            'urls': []
        })
    
    return emails

# Generate Hybrid V-Triad
hybrid_vtriad_emails = generate_hybrid_vtriad_emails(50)
print(f"\n✓ Generated {len(hybrid_vtriad_emails)} Hybrid V-Triad emails")
print(f"  Sample sender: {hybrid_vtriad_emails[0]['sender']}")
print(f"  Sample subject: {hybrid_vtriad_emails[0]['subject']}")
````

````python
def save_regenerated_datasets(plain_llm_emails, hybrid_vtriad_emails):
    """
    Save regenerated datasets as JSON files for integration
    """
    from pathlib import Path
    import json
    
    output_dir = Path('data/regenerated_datasets')
    output_dir.mkdir(parents=True, exist_ok=True)
    
    # Save to JSON files
    plain_path = output_dir / 'plain_llm_regenerated.json'
    hybrid_path = output_dir / 'hybrid_vtriad_regenerated.json'
    
    with open(plain_path, 'w') as f:
        json.dump(plain_llm_emails, f, indent=2)
    
    with open(hybrid_path, 'w') as f:
        json.dump(hybrid_vtriad_emails, f, indent=2)
    
    print(f"✓ Saved Plain LLM to: {plain_path}")
    print(f"✓ Saved Hybrid V-Triad to: {hybrid_path}")
    
    return plain_path, hybrid_path

# Save the generated datasets
plain_path, hybrid_path = save_regenerated_datasets(plain_llm_emails, hybrid_vtriad_emails)

print("\n" + "="*80)
print("GENERATION COMPLETE")
print("="*80)
print(f"✓ Plain LLM: {len(plain_llm_emails)} emails with {sum(1 for e in plain_llm_emails if e['urls'])} URLs")
print(f"✓ Hybrid V-Triad: {len(hybrid_vtriad_emails)} emails with {sum(1 for e in hybrid_vtriad_emails if e['urls'])} URLs")
print(f"\nNext: Run Phase 2 normalization to extract URLs and create new master_emails.csv")
````
